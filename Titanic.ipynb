```python
import pandas as pd
import matplotlib.pyplot as plt
```



In dit bestand zal worden gekeken in hoever "Vrouwen en kinderen eerst" gold bij deze selectie van de passagies van de Titanic, en dus hoeveel van de overlevenden vrouwen en kinderen (<18) zijn. Hieronder volgt eerst de hele dataset.

```python
titanic = pd.read_csv("titanic.csv")
print(titanic)
```

```python
print(titanic.info())
```



Dit zijn de overlevenden:

```python
overleefd = titanic[titanic["Survived"] == 1]
print(overleefd)
```



Dit zijn de overledenen:

```python
overleden = titanic[titanic["Survived"] == 0]
print(overleden)
```

Hieronder wordt de tabel gesplitst in overleden en overlevende mannen, vrouwen en kinderen.

```python
overleefdmannen = overleefd[(overleefd["Sex"] == "male") & (overleefd["Age"] >= 18)]
overleefdvrouwen = overleefd[(overleefd["Sex"] == "female") & (overleefd["Age"] >= 18)]
overleefdkinderen = overleefd[overleefd["Age"] <= 18]
overledenmannen = overleden[(overleden["Sex"] == "male") & (overleden["Age"] >= 18)]
overledenvrouwen = overleden[(overleden["Sex"] == "female") & (overleden["Age"] >= 18)]
overledenkinderen = overleden[overleden["Age"] <= 18]
```

```python
print("Aantal overlevende mannen: " + str(len(overleefdmannen)))
print("Aantal overlevende vrouwen: " + str(len(overleefdvrouwen)))
print("Aantal overlevende kinderen: " + str(len(overleefdkinderen)))
print("")
print("Aantal overleden mannen: " + str(len(overledenmannen)))
print("Aantal overleden vrouwen: " + str(len(overledenvrouwen)))
print("Aantal overleden kinderen: " + str(len(overledenkinderen)))
print("")
print("Totaal aantal mannen: " + str(len(overleefdmannen) + len(overledenmannen)))
print("Totaal aantal vrouwen: " + str(len(overleefdvrouwen) + len(overledenvrouwen)))
print("Totaal aantal kinderen: " + str(len(overleefdkinderen) + len(overledenkinderen)))
print("")
print("Aantal overlevenden: " + str(len(overleefd)))
print("Aantal overledenen: " + str(len(overleden)))
print("")
print("Percentages binnen de overlevenden:")
print("Overlevende mannen: " + str(round(len(overleefdmannen) / len(overleefd) * 100, 2)) + " %")
print("Overlevende vrouwen: " + str(round(len(overleefdvrouwen) / len(overleefd) * 100, 2)) + " %")
print("Overlevende kinderen: " + str(round(len(overleefdkinderen) / len(overleefd) * 100, 2)) + " %")
print("")
print("Percentages binnen de overledenen:")
print("Overleden mannen: " + str(round(len(overledenmannen) / len(overleden) * 100, 2)) + " %")
print("Overleden vrouwen: " + str(round(len(overledenvrouwen) / len(overleden) * 100, 2)) + " %")
print("Overleden kinderen: " + str(round(len(overledenkinderen) / len(overleden) * 100, 2)) + " %")
print("")
print("Overlevingskans:")
totaalmannen = len(overleefdmannen) + len(overledenmannen)
totaalvrouwen = len(overleefdvrouwen) + len(overledenvrouwen)
totaalkinderen = len(overleefdkinderen) + len(overledenkinderen)
print("Overlevingskans mannen: " + str(round(len(overleefdmannen) / totaalmannen * 100, 2)) + " %")
print("Overlevingskans vrouwen: " + str(round(len(overleefdvrouwen) / totaalvrouwen * 100, 2)) + " %")
print("Overlevingskans kinderen: " + str(round(len(overleefdkinderen) / totaalkinderen * 100, 2)) + " %")
```

Het ziet ernaar uit dat binnen deze dataset de overlevingskans voor vrouwen het hoogste is, daarna voor kinderen en ten slotte is deze het laagst voor mannen (waar er wel meer dan twee keer zoveel van zijn dan vrouwen, maar meer dan de helft van de overlevenden is vrouw). Uiteraard gaat deze dataset niet over alle passagiers (2223, de dataset gaat over 887 passagiers), maar het lijkt erop dat het voor deze keer een voordeel is om een vrouw te zijn in 1912 (en natuurlijk hoef je dan over een paar jaar niet te vechten in de Eerste Wereldoorlog, of als je in 1912 kind bent in de Tweede). "Vrouwen en kinderen eerst" is dus ook echt zo in levensbedreigende situaties.
